```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
x_m = 1  # Scale parameter
alpha = 1.5  # Shape parameter (1 < alpha <= 2 for finite mean, infinite variance)
true_mean = 3

# Generate samples
samples = (np.random.pareto(alpha, 100000) + 1) * x_m

# Plot the histogram
plt.hist(samples, bins=50, density=True, alpha=0.6, color='blue', edgecolor='black')
plt.title(f"Pareto Distribution (x_m={x_m}, alpha={alpha})")
plt.xlabel("x")
plt.ylabel("Density")
plt.show()

# Compute the sample mean
sample_mean = np.mean(samples)
print(f"Sample mean: {sample_mean}")
```

```python
sample_sizes = np.arange(1,6001)

absolute_errors = [abs(np.mean(samples[ :n]) - true_mean) for n in sample_sizes]

plt.figure(figsize=(10,6))
plt.plot(sample_sizes, absolute_errors)
plt.title('Error Convergence')
plt.xlabel('sample error')
plt.ylabel('Absolute Error')
plt.grid(True)
plt.show()
```
