```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function f(x)
def f(x):
    return (2**x) / (x**2 + 1)

# Define the range for x, avoiding division by zero
x_min = -10
x_max = 10

# Set up rejection sampling
samples = []
num_samples = 1000
max_y = f(x_max)  # Upper bound for rejection sampling

np.random.seed(0)  # For reproducibility

while len(samples) < num_samples:
    # Generate random x in the specified range
    x = np.random.uniform(x_min, x_max)
    # Generate a random y in [0, max_y]
    y = np.random.uniform(0, max_y)
    # Accept x if y < f(x)
    if y < f(x):
        samples.append(x)

# Convert the list to a numpy array for easy handling
samples = np.array(samples)

# Plot the sampled points
plt.figure(figsize=(10, 6))
plt.hist(samples, bins=20, density=True, alpha=0.6, color="skyblue", edgecolor="black")
x_vals = np.linspace(x_min, x_max, 500)
plt.plot(x_vals, f(x_vals), "r-", lw=2, label="f(x) = (2^x) / (x^2 + 1")
plt.xlabel("x")
plt.ylabel("Density")
plt.legend()
plt.title("Sampling from f(x) = (2^x) / (x^2 + 1)")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function f(x) = 2^x / (x^2 + 1)
def f(x):
    return (1) / np.pi*(x**2 + 1)

# Define the range for x
x_min = -10000  # Adjust this as needed for your sampling range
x_max = 10000

# Set up rejection sampling
samples = []
num_samples = 100
max_y = max(f(np.linspace(x_min, x_max, 1000)))  # Upper bound for rejection sampling

np.random.seed(0)  # For reproducibility

while len(samples) < num_samples:
    # Generate random x in the specified range
    x = np.random.uniform(x_min, x_max)
    # Generate a random y in [0, max_y]
    y = np.random.uniform(0, max_y)
    # Accept x if y < f(x)
    if y < f(x):
        samples.append(x)

# Convert the list to a numpy array
samples = np.array(samples)

# Compute the mean of the sampled points
mean_sampled_points = np.mean(samples)

# Display the mean
print("Mean of the sampled points:", mean_sampled_points)

# Plot the sampled points and the function
plt.figure(figsize=(10, 6))
plt.hist(samples, bins=20, density=True, alpha=0.6, color="skyblue", edgecolor="black", label="Sampled Points")
x_vals = np.linspace(x_min, x_max, 500)
plt.plot(x_vals, f(x_vals), "r-", lw=2, label=r"$f(x) = \frac{2^x}{x^2 + 1}$")
plt.xlabel("x")
plt.ylabel("Density")
plt.legend()
plt.title("Sampling from f(x) = 2^x / (x^2 + 1)")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import cauchy

# Parameters
x_0 = 0  # Location (median)
gamma = 1  # Scale

# Generate values from the Cauchy distribution
x = np.linspace(-10, 10, 1000)
y = cauchy.pdf(x, loc=x_0, scale=gamma)

# Plot the PDF
plt.plot(x, y, label="Cauchy Distribution")
plt.title("Cauchy Distribution with Finite Mean but Infinite Variance")
plt.xlabel("x")
plt.ylabel("Probability Density")
plt.legend()
plt.show()
```

```python
import numpy as np
from scipy.stats import cauchy

# Parameters
x_0 = 0  # Location (median)
gamma = 5  # Scale
num_samples = 1000  # Number of random samples

# Sample 100 points from the Cauchy distribution
samples = cauchy.rvs(loc=x_0, scale=gamma, size=num_samples)

# Compute the mean of the samples
mean_samples = np.mean(samples)

# Print the result
print(f"Mean of the 100 samples: {mean_samples}")
```

The above (Cauchy) test confirm that, even after increasing the number of smample points, the sample mean does not converge to any particular point. This shows that SLLN does not hold here.
