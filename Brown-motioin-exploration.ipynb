We try to understand the definition of Brownian motion here.

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
T = 1.0  # Total time
N = 100  # Number of time steps
dt = T / N  # Time step size
t = np.linspace(0, T, N+1)  # Time points

# Simulate Brownian motion
dB = np.sqrt(dt) * np.random.randn(N)  # Increments (Gaussian with mean 0, variance dt)
B = np.concatenate([[0], np.cumsum(dB)])  # Brownian motion path

# Plot Brownian motion
plt.figure(figsize=(10, 6))
plt.plot(t, B, label="Brownian Motion Path")
plt.title("Brownian Motion Simulation")
plt.xlabel("Time")
plt.ylabel("B(t)")
plt.axhline(0, color='black', linestyle='--', linewidth=1)
plt.legend()
plt.grid()
plt.show()
```
