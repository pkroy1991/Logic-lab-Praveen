```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def reimann_sum(f, a, b, n):
    dt = (b-a)/n
    x = np.linspace(a, b-dt, n)
    my_sum = np.sum(f(x)*dt)
    return my_sum, dt
```

```python
actual_value = 5.3
```

```python
def f(x):
    return -(x**2) + (2*(x**3))
```

```python
errors = []
dts = []
n_values = 100
```

```python
for n in range(1, n_values):
    my_sum, dt = reimann_sum(f, 0, 2, n)
    error = abs(actual_value - my_sum)
    errors.append(error)
    dts.append(dt)
```

```python
inverse_dts = [2/(dt**(1/2)) for dt in dts]
```

```python
plt.loglog(errors, dts, marker = 'o', linestyle = '-', color = 'b', label = 'Error = Actual- Riemann sum')
plt.loglog(dts, inverse_dts, marker = 'o', linestyle = '-', color = 'r', label = '1/dt')
plt.xlabel(" width of intervel")
plt.ylabel("error")
plt.title("Error vs dt with comparison to 1/dt")
plt.legend()
#plt.grid(True)
plt.show()
```

```python
def f(x):
    return -(x**2) + (2*(x**3))
```

```python
n = 1000
rand_points = []
for _ in range(n):
        rand_points.append(np.random.uniform(0, 2))
```

```python
all_sum = 0
for x in rand_points:
        all_sum = all_sum + f(x)
avg_sum = all_sum/n
```

```python
print(avg_sum)
```

```python
actual_value = -(8/3) + 8
error = actual_value - avg_sum
```

```python
print(error)
```

```python
def mc_intgr(f, a, b, n):
    x = []
    rand_point = 0
    all_sum = 0
    for _ in range(n):
        rand_point = np.random.uniform(a, b)
        all_sum = all_sum + f(rand_point)
    avg_sum = (b-a)*(all_sum/n)
    return avg_sum
```

```python
a = 0
b = 2
n_values = 10000
```

```python
mc_intgr(f, a, b, 10000)-actual_value
```

```python
errors = []
avg_sums = []
error_estimator = []
all_n_values = []
for i in range(5, n_values,20):
    r = mc_intgr(f, a, b, i)
    avg_sums.append(r)
    error = abs(r - actual_value)
    errors.append(error)
    error_estimator.append(2/i**(1/2))
    all_n_values.append(i)
```

```python
plt.loglog(all_n_values, errors, marker = 'o', linestyle = '-', color = 'b', label = 'Error = Actual- Expectated sum')
plt.loglog(all_n_values, error_estimator, marker = 'o', linestyle = '-', color = 'r', label = 'Sigma/sqrt(n)')
plt.xlabel('avg sum')
plt.ylabel('all n values')
plt.legend()
plt.show()
```
