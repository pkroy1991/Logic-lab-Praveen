### Area under the curve : 
Let's consider the area under the curve $f(t)$ from $S$ to $T$ = $\int_{S}^{T} f(t) dt$.

- **Sum of area elements**: The idea is to divide the interval $[S,T]$ into multiple (say $n$) smaller intervals: $[t_1, t_2, \cdots, t_{n+1}]$ with $t_0 = S$ and $t_{n+1} = T$ and approximate the area of the smaller (approximately rectangular) area elements $(\Delta A)_i$. The total area is then approximated to be the sum of these area elements $A = \sum (\Delta A)_i$.

- Note that, we can have various ways to approximate the smaller area elements:
  - **Left**: $(\Delta A)_i = f(t_i) * (\Delta t)_i$ where $(\Delta t)_i = t_{i+1} - t_i$
  - **Right**: $(\Delta A)_i = f(t_{i+1}) * (\Delta t)_i$
  - **Central**: $(\Delta A)_i = \frac{1}{2} \left\{ f(t_{i}) + f(t_{i+1}) \right\} * (\Delta t)_i$

    &nbsp;  
- In the limit, all these methods will converge to the exact area. However, the rate of convergence can be different for different schemes as we will see soon.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def left_rule(f, a, b, n):
    x = np.linspace(a, b, n+1)
    y = f(x)
    h = (b - a) / n
    area = np.sum(y[0:-1]) * h
    return area
    
def right_rule(f, a, b, n):
    x = np.linspace(a, b, n+1)
    y = f(x)
    h = (b - a) / n
    area = np.sum(y[1:]) * h
    return area
    
def trapezoidal_rule(f, a, b, n):
    x = np.linspace(a, b, n+1)
    y = f(x)
    h = (b - a) / n
    area = (y[0] + 2 * np.sum(y[1:-1]) + y[-1]) * h / 2
    return area
```

- **Example Integral**: Let us find area under the curve $f(t) = t^3 - t^2 +t$ from $t = -5$ to $10$ with $F(t) = \int f(t) dt = \frac{1}{4} t^4 - \frac{1}{3} t^3 + \frac{1}{2} t^2$.

```python
def f(t):
    return t**3 - t*t + t
def F(t):
    return t**4/4 - t**3/3 + t*t/2
def left_area(n):
    return left_rule(f,-5,10,n)
def right_area(n):
    return right_rule(f,-5,10,n)
def central_area(n):
    return trapezoidal_rule(f,-5,10,n)
def h(n):
    return 15 / n

def exact_area():
    return F(10) - F(-5)

def left_values(r):
    return [(h(r*i), left_area(r*i)) for i in range(1,10)]
def right_values(r):
    return [(h(r*i), right_area(r*i)) for i in range(1,10)]
def central_values(r):
    return [(h(r*i), central_area(r*i)) for i in range(1,10)]

print("Exact Area = ", exact_area())
mult=10
# Plot the solution
plt.plot(*zip(*left_values(mult)), color='r', label='Left')
plt.plot(*zip(*central_values(mult)), color='g', label='Central')
plt.plot(*zip(*right_values(mult)), color='b', label='Right')
plt.xlabel('h')
plt.ylabel('Area')
plt.title('Numerical Integration of ODE')
plt.legend()
plt.grid(True)
plt.show()

def left_errors(r):
    return [(h(r*i), -exact_area()+left_area(r*i)) for i in range(1,10)]
def right_errors(r):
    return [(h(r*i), -exact_area()+right_area(r*i)) for i in range(1,10)]
def central_errors(r):
    return [(h(r*i), -exact_area()+central_area(r*i)) for i in range(1,10)]

# Plot the errors
plt.plot(*zip(*left_errors(mult)), color='r', label='Left')
plt.plot(*zip(*central_errors(mult)), color='g', label='Central')
plt.plot(*zip(*right_errors(mult)), color='b', label='Right')
plt.xlabel('h')
plt.ylabel('Error in Area')
plt.title('Error in various schemes for polynomial')
plt.legend()
plt.grid(True)
plt.show()
```

# Numerical Errors and Convergence Rates for different schemes

Different numerical schemes produce errors charasteristic to them. Understanding these errors and the behavior of convergence is crucial for effectively using these methods.

## Trapezoidal Rule Recap
The Trapezoidal Rule approximates the integral of a function $f(x)$ over the interval $[a, b]$ by dividing the interval into $n$ subintervals and summing the areas of the resulting trapezoids:

$\int_a^b f(x) \, dx \approx \frac{b - a}{2n} \left[ f(x_0) + 2 \sum_{i=1}^{n-1} f(x_i) + f(x_n) \right] $

where $ x_i = a + i \frac{b-a}{n} $ for $ i = 0, 1, \ldots, n $.

## Numerical Error
### Truncation Error
The error introduced by approximating the integral using the Trapezoidal Rule is known as truncation error. For the Trapezoidal Rule, the truncation error in one subinterval is proportional to the second derivative of the function being integrated. The error for the entire interval can be expressed as:

$ E_T = -\frac{(b - a)^3}{12n^2} f''(\xi) $

for some $\xi$ in $[a, b]$.

### Key Points:
1. **Dependence on \( f''(x) \)**: The error depends on the second derivative of \( f(x) \). If \( f(x) \) is nearly linear (small \( f''(x) \)), the error will be smaller.
2. **Proportional to $\left( \frac{1}{n^2} \right)$**: The error decreases quadratically with the number of subintervals \( n \). Doubling the number of intervals reduces the error by a factor of 4.

## Convergence
Convergence refers to how the approximation improves as the number of subintervals \( n \) increases.

### Error Analysis:
The global error for the Trapezoidal Rule, when summing over all subintervals, can be approximated as:

$ E_T \approx -\frac{(b - a)^3}{12n^2} f''(\xi) $

### Convergence Rate:
The Trapezoidal Rule converges with an error that is $\mathcal{O}\left(\frac{1}{n^2}\right)$. This quadratic convergence rate means that as $n$ increases, the error decreases rapidly. Specifically, if you double the number of subintervals, the error is reduced by a factor of 4.

## Local truncation error

 - We assume that $y(t)$ is a well-behaved function, and hence we can use Taylor's theorem.
 - With this, $y(t+\Delta t) = y(t) + \dot y \times \Delta t + \ddot y \times \frac{(\Delta t )^2}{2} + \dddot y \times \frac{(\Delta t )^3}{6} + \cdots$
 - The integration step in Euler Scheme (and similarly for the ), $y(t+\Delta t) = y(t) + \dot y \times \Delta t$ obviously has the local truncation error of $\mathcal{O}(\Delta t^2)$.
