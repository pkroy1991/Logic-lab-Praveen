```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Set the number of trials
num_trials =50000

# Simulate coin flips: 1 for heads, 0 for tails
# Assuming a fair coin, the probability of heads (1) is 0.5
flips = np.random.choice([0, 1], size=num_trials)

# Compute cumulative average of heads at each trial
cumulative_average = np.cumsum(flips) / np.arange(1, num_trials + 1)

# Plot the results
plt.figure(figsize=(10, 6))
plt.plot(cumulative_average, label='Cumulative Average of Heads')
plt.axhline(y=0.5, color='r', linestyle='--', label='Expected Value (0.5)')
plt.xlabel('Number of Trials')
plt.ylabel('Cumulative Average')
plt.title('Law of Large Numbers: Coin Flip Simulation')
plt.legend()
plt.show()
```

```python
def random_function_normal(num_samples):
    data = np.random.normal(10, 100, num_samples)
    return np.mean(data)
```

```python
random_function_normal(100000)
```

```python
def plot_hist_random(rand_func, num_samples):
    # Generate samples by repeatedly calling the random_function
    samples = [rand_func(num_samples) for _ in range(10000)]

    print(np.std(samples))
    
    # Plot the histogram of the samples
    plt.figure(figsize=(8, 6))
    plt.hist(samples, bins=100, color='skyblue', edgecolor='black', density=True, alpha=0.7)
    
    # Add labels and title
    plt.xlabel('Value')
    plt.ylabel('Probability Density')
    plt.title('Histogram of Samples from Random Function')
    
    # Show the plot
    plt.show()
```

```python
plot_hist_random(random_function_normal, 10)
plot_hist_random(random_function_normal, 100)
plot_hist_random(random_function_normal, 500)
```

```python
def random_function_coin_toss(num_samples):
    data = np.random.choice([0, 1], size=num_samples)
    return np.mean(data)
```

```python
plot_hist_random(random_function_coin_toss, 10)
plot_hist_random(random_function_coin_toss, 100)
plot_hist_random(random_function_coin_toss, 500)
plot_hist_random(random_function_coin_toss, 5000)
```
