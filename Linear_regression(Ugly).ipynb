```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
def f(x, m, c, e):
    return m*x + c + e
```

```python
m = 1
c = 2
mu = 0
sigma = .05
```

```python
n = 10
rand_points = []
rand_point = 0
for _ in range(n):
    rand_point = np.random.uniform(0, n)
    rand_points.append(rand_point)
```

```python
rand_val_normal = []
for _ in range(n):
    rand_val_normal.append(np.random.normal(mu, sigma))
```

```python
f_values = []
f_val = 0
for i in range(n):
    f_val = f(rand_points[i], m, c, rand_val_normal[i])
    f_values.append(f_val)
```

```python
plt.plot(rand_points, f_values, marker = 'o', linestyle = '', color = 'b', label = 'values of f')
plt.xlabel('Random points')
plt.ylabel('Random values')
plt.show()
```

```python
N = 10
m = []
errors_m = []
error_m = 0
for j in range(N):
    for i in range(n):
        m_i = abs((f_values[i] - c))/rand_points[i]
        m.append(m_i)
    m_estimator = sum(m)/n
    error_m = abs(m_estimator - 1)
    errors_m.append(error_m)
print(errors_m)
```

```python
plt.figure(figsize=(10,6))
plt.plot(rand_points, f_values, marker = 'o', linestyle = '', color = 'b', label = 'values of f')
plt.xlabel('Random points')
plt.ylabel('Random values')
plt.show()
```

```python
n_values = []
for i in range(n):
    n_values.append(i)
```

```python
plt.plot(n_values, errors_m, marker = 'o', linestyle = '', color = 'r', label = 'error in m')
plt.show()
```
