```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define parameteres
true_mean = 0.5 #Expected value for the uniform [0,1]
np.random.seed(42) # For reproducibility
```

```python
# Generate random samples

sample_sizes = np.arange(1,10001)

# Sample sizes from 1 to 10,000

samples = np.random.uniform(0,1,max(sample_sizes)) # Random samples

# Compute running sample means
sample_means = [np.mean(samples[ :n]) for n in sample_sizes]

# Plot the convergence

plt.figure(figsize = (10,6))
plt.plot(sample_sizes, sample_means, label = 'Sample mean')
plt.axhline(y=true_mean, color='r', linestyle='--', label='True Mean')
plt.title('Law of large numbers')
plt.xlabel('Sample Sizes')
plt.ylabel('Sample Mean')
plt.legend()
plt.grid(True)
plt.show()
```

```python
absolute_errors = [abs(np.mean(samples[ :n]) - true_mean) for n in sample_sizes]

plt.figure(figsize=(10,6))
plt.plot(sample_sizes, absolute_errors)
plt.title('Error Convergence')
plt.xlabel('sample error')
plt.ylabel('Absolute Error')
plt.grid(True)
plt.show()
```
