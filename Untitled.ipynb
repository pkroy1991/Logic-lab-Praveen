```python
def solve_ode(n):
    x = 1
    y = 2
    t = 0
    t0 = 0
    t1 = 10
    dt = (t1 - t0)/n
    for i in range(n):
        t = t + dt
        y = y + (y + x)*dt
        x = x + (np.sin(x) + np.cos(y))*dt
    return (x,y,t)
```
